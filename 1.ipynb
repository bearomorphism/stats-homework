```python
def preprocess(s):
    return sorted(list(map(int, s.split())))

v1 = preprocess("20 12 39 38 41 43 51 52 59 55 53 59 50 58 35 38 23 32 43 53")
v2 = preprocess("18 45 62 59 53 25 13 57 42 55 56 38 41 36 50 62 45 55")
```

```python
def stem(arr):
    dic = defaultdict(list)
    for i in arr:
        dic[i // 10].append(i % 10)
    for k, v in dic.items():
        print(k, end=': ')
        print(' '.join(map(str, v)))

print("Variety 1")
stem(v1)
print("Variety 2")
stem(v2)
```
